```python
import os

def get_json_files(folder_path):
    json_files = []
    for file in os.listdir(folder_path):
        if file.endswith('.json'):
            json_files.append(os.path.join(folder_path, file))
    return json_files

folder_path =  r"C:\Users\I4324\Downloads\ipl_json"
json_files = get_json_files(folder_path)
```

```python
def overs_sheet(data):
    df = pd.DataFrame(columns=['Over', 'Batter', 'Bowler', 'Non-Striker', 'Batter Runs', 'Extras', 'Total Runs', 'Wicket Kind', 'Player Out', 'Fielders', 'match_date', 'match_btwn', 'venue', 'season', 'result', 'player_of_match', 'toss', 'innings'])

    info = data['info']
    innings_data = data['innings']
    x = 0
    for inning in innings_data:
        total_runs_counter = 0
        ings = 'First_inngs' if x == 0 else 'Second_inngs'
        for over in inning['overs']:
            for delivery in over['deliveries']:
                over_number = over['over']
                batter = delivery['batter']
                bowler = delivery['bowler']
                non_striker = delivery['non_striker']
                batter_runs = delivery['runs']['batter']
                extras = delivery['runs'].get('extras', 0)
                total_runs = total_runs_counter + batter_runs + extras
                total_runs_counter = total_runs

                wicket_kind = None
                player_out = None
                fielders = None

                if 'wickets' in delivery:
                    #print(delivery)
                    for wicket in delivery['wickets']:
                        wicket_kind = wicket['kind']
                        player_out = wicket['player_out']
                        if wicket_kind not in ['bowled', 'lbw','caught and bowled','run out','retired hurt','hit wicket']:
                            fielders = ', '.join([fielder['name'] for fielder in wicket['fielders']])
                        else:
                            pass

                match_date = '-'.join(info['dates'])
                match_btwn = f"{info['teams'][0]} VS {info['teams'][1]}"
                venue = info['venue']
                season = info['season']
                try:
                    if info['outcome']['result'] == 'tie':
                        result = f"The match between {info['outcome']['eliminator']} ended in a tie with no winner determined."
                except:
                    try:
                        result = f"{info['outcome']['winner']} Won  by {info['outcome']['by']['runs']} runs"
                    except:
                        result = f"{info['outcome']['winner']} Won  by {info['outcome']['by']['wickets']} wickets"
                player_of_match = str(info['player_of_match'])
                toss = f"Toss Won by {info['toss']['winner']} And Decided To {str.upper(data['info']['toss']['decision'])} First"
                try:
                    match_number=str(info['event']['match_number'])
                except:
                    match_number=str(info['event']['stage'])

                df = df.append({'Over': over_number, 'Batter': batter, 'Bowler': bowler, 'Non-Striker': non_striker,
                                'Batter Runs': batter_runs, 'Extras': extras, 'Total Runs': total_runs,
                                'Wicket Kind': wicket_kind, 'Player Out': player_out, 'Fielders': fielders,
                                'match_date': match_date, 'match_btwn': match_btwn, 'venue': venue,
                                'season': season, 'result': result, 'player_of_match': player_of_match,
                                'toss': toss, 'innings': ings,'match_number':match_number}, ignore_index=True)

        x += 1

    df.to_excel(f"C:/Users/I4324/Downloads/IPL DATA/{'-'.join(info['dates'])}_{match_number}_{season}.xlsx", index=False)
    print("Data saved to the  respective folder")

```
